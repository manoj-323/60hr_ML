# plotting a vector

```python
plt.quiver(0,0,5,5)
plt.show()
```

```python
plt.quiver(0,0,5,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,5,-5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```

# substraction of two vectors

```python
vector_3 = np.asarray([0,0,2,3])
vector_4 = np.asarray([0,0,3,-2])
difference = vector_3-vector_4
print(difference)
```

# dot product of two vectors :-
    -> gives a scaler quantity
    -> np.dot(vector_1,vector_2) function is used to calculate dot product

```python
vector_9 = np.asarray([0,0,2,3])
vector_10 = np.asarray([0,0,4,4])
dot_product = np.dot(vector_9,vector_10)
print(vector_9)
print(vector_10)
print(dot_product)
```

# cross product of 2 vectors :-
    -> it gives a vector quantity
    -> np.cross(vector_1,vector_2) is the function used to calsulate the cross product

```python
vector_11 = np.array([2,3,3])
vector_12 = np.array([4,4,4])
cross_product = np.cross(vector_11,vector_12)
print(vector_11)
print(vector_12)
print(cross_product)
```

# projection of one vector on another vector:-

```python
vector_13 = np.asarray([2,5])
vector_14 = np.asarray([8,-6])
# magnitude of vector_14
magnitude_vector_14 = np.sqrt(np.sum(vector_14**2))
# dot product of two vectors
dot_product_of_two = np.dot(vector_13,vector_14)
# now
proj_13_on_14 = (dot_product_of_two/magnitude_vector_14**2)*vector_14
```

```python
print(vector_13)
print(vector_14)
print('Projection: ', proj_13_on_14)
```
