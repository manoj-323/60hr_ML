```python
# house price dataset
house = pd.read_csv(r"C:\Users\22213\anaconda3\Lib\site-packages\sklearn\datasets\data\boston_house_prices.csv")
```

```python
house.info
```

```python
house.columns = house.iloc[0]
house=house[1:]
```

```python
house.rename(columns={'MEDV':'PRICE'}, inplace=True)
```

```python
house.head()
```

```python
# distplot
sns.displot(house['PRICE'], stat='density', kde=True)
```

```python
# co-relation matrix
#  -positive co-relation
#  -negative co-relation
```

```python
corelation = house.corr()
```

```python
# heat map
plt.figure(figsize=(10,10))
sns.heatmap(corelation, cbar=True, square=True, fmt='.1f', annot=True,
           annot_kws={'size':8}, cmap='Blues') 
```
