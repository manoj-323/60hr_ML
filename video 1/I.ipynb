```python
# dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
sonar_dataset = pd.read_csv(r"C:\Users\22213\OneDrive\Desktop\CSV\sonar data.csv", header = None)
```

```python
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(sonar_dataset[60])
sonar_dataset['target'] = labels
```

```python
sonar_dataset.sample(5)
```

```python
# seprerating data and labels
X = sonar_dataset.drop(columns=[60,'target'], axis=1)
Y = sonar_dataset['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .2, random_state = 2, stratify = Y)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

# Model training:
    logistic regression

```python
model = LogisticRegression()
```

```python
# training model
model.fit(X_train,Y_train)
```

**Model evaluation**

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data: ',test_data_accuracy)
```
